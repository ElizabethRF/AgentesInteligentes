```python
import urllib.request as ur
import re 

f = ur.urlopen('https://www.aireuropa.com/es-es/ofertas-de-vuelos-desde-madrid')
s = f.read().decode()
print(s)

#ss = split('<div class = "col-sm-4"')
#vuelos = re.findall( "<div class=\"col-sm-4\".?*<span class=\"fare-atom-origin-destination-origin-city\".?*>(.?*)</span>.?*<span class=\"fare-atom-origin-destination-connector\".?*>(.?*)</span>.?*<span class=\"fare-atom-origin-destination-destination-city\".?*>(.?*)</span>.?*(dd/dd/dddd).?*(dd/dd/dddd).?*<span class=\"fare-atom-price-currency-code\".?*>(.?*)</span>.?*<span class=\"fare-atom-price-total-price\".?*>(.?*)</span>.?*<span class=\"fare-atom-journey-type\".?*>(.?*)</span>", s, re.DOTALL)
f.close()
```

```python
# Método para eliminar tags html que no nos interesan con ER 
def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr,'',raw_html)
    return cleantext
```
